```python
# Sistema Bancário em Python
# Autor: Gabriel Pessine
# Data: 21/09/2024
# Descrição: Sistema Bancário simples em Python, para o desafio do DIO/NTT - Modulo Sintaxe Basica com Pyhton

saldo = 0  # Saldo atual da conta
depositos = []  # Lista para armazenar os valores dos depósitos
saques = []  # Lista para armazenar os valores dos saques
limite_saques_diarios = 3  # Limite máximo de saques por dia
valor_maximo_saque = 500  # Valor máximo por saque
numero_saques = 0  # Contador de saques realizados no dia

def depositar(valor):
    """Realiza um depósito na conta bancária.

    Args:
        valor (float): Valor a ser depositado.
    """
    if valor > 0:
        global saldo  # Acessa a variável global saldo
        saldo += valor
        depositos.append(valor)
        print("Depósito realizado com sucesso.")
    else:
        print("Valor inválido. O valor do depósito deve ser positivo.")

def sacar(valor):
    """Realiza um saque na conta bancária.

    Args:
        valor (float): Valor a ser sacado.
    """
    global saldo, numero_saques  # Acessa as variáveis globais saldo e numero_saques
    if numero_saques < limite_saques_diarios and valor <= valor_maximo_saque and saldo >= valor:
        saldo -= valor
        saques.append(valor)
        numero_saques += 1
        print("Saque realizado com sucesso.")
    else:
        if numero_saques >= limite_saques_diarios:
            print(f"Você atingiu o limite máximo de {limite_saques_diarios} saques diários.")
        elif valor > valor_maximo_saque:
            print(f"O valor do saque excede o limite máximo por operação, que é de R$ {valor_maximo_saque:.2f}.")
        else:
            print(f"Saldo insuficiente. Seu saldo atual é de R$ {saldo:.2f}.")

def extrato():
    """Exibe o extrato da conta bancária, mostrando todos os depósitos e saques realizados, além do saldo atual."""
    print("\n==========")
    print("Extrato:")
    print("==========\n")
    if not depositos and not saques:
        print("Não foram realizadas movimentações.")
    else:
        for deposito in depositos:
            print(f"Depósito: R$ {deposito:.2f}")
        for saque in saques:
            print(f"Saque: R$ {saque:.2f}")
        print(f"Saldo: R$ {saldo:.2f}")
    print("=" * 30)

# Loop principal do programa
while True:
    print("\nOperações:\n1 - Depósito\n2 - Saque\n3 - Extrato\n4 - Sair")
    opcao = int(input("Digite a opção desejada: "))

    if opcao == 1:
        valor = float(input("Digite o valor a ser depositado: "))
        depositar(valor)
    elif opcao == 2:
        valor = float(input("Digite o valor a ser sacado: "))
        sacar(valor)
    elif opcao == 3:
        extrato()
    elif opcao == 4:
        break
    else:
        print("Opção inválida.")
```
