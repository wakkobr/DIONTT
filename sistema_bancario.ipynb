```python
# Sistema Bancário em Python
# Autor: Gabriel Pessine
# Data: 21/09/2024
# Descrição: Sistema Bancário simples em Python, para o desafio do DIO/NTT - Modulo Sintaxe Basica com Pyhton

saldo = 0            # Saldo atual da conta
depositos = []       # Lista para armazenar os depósitos
saques = []          # Lista para armazenar os saques
lim_dia_saque = 3    # Limite máximo de saques por dia
val_max_saque = 500  # Valor máximo por saque
cont_saque = 0       # Contador de saques realizados no dia

# Função DEPOSITAR

def depositar(valor):
    # Realiza depósito na conta.
    # Var: valor (float): valor do depósito

    if valor > 0:
        global saldo
        saldo += valor
        depositos.append(valor)
        print("\n==================================================")
        print(f"Depósito de R$ {valor:.2f} realizado com sucesso.")
        print("==================================================")
    else:
        print("\n==================================================")
        print("Valor inválido, deve ser maior que zero.")
        print("==================================================")

# Função SACAR

def sacar(valor):
    # Realiza saque da conta.
    # Var: valor (float): valor do saque

    global saldo, cont_saque
    if cont_saque < lim_dia_saque and valor <= val_max_saque and saldo >= valor and valor > 0:
        saldo -= valor
        saques.append(valor)
        cont_saque += 1   # Incrementa contador de saques, para checar o limite depois
        print("\n==================================================")
        print(f"Saque de R$ {valor:.2f} realizado com sucesso.")
        print("==================================================")
    else:
        if cont_saque >= lim_dia_saque:
            print("\n==================================================")
            print(f"Você atingiu o limite máximo de {lim_dia_saque} saques diários.")
            print("==================================================")
        elif valor > val_max_saque:
            print("\n=============================================================================")
            print(f"O valor do saque excede o limite máximo por operação, que é de R$ {val_max_saque:.2f}.")
            print("=============================================================================")
        elif valor <= 0:
            print("\n==================================================")
            print("Valor inválido, deve ser maior que zero.")
            print("==================================================")
        else:
            print("\n==================================================")
            print(f"Saldo insuficiente. Seu saldo atual é de R$ {saldo:.2f}.")
            print("==================================================")

# Função EXTRATO

def extrato():
    # Mostra o extrato completo da conta

    print("\n==================================================")
    print("Extrato:")
    print("==================================================\n")
    if not depositos and not saques:
        print("\n==================================================")
        print("Não foram realizadas movimentações nessa conta.")
        print("==================================================")
    else:
        for deposito in depositos:
            print(f"Depósito: R$ {deposito:.2f}")
        for saque in saques:
            print(f"Saque: R$ {saque:.2f}")
        print("\n=========================")
        print(f"Saldo atual: R$ {saldo:.2f}")
        print("=========================")

# Programa principal
while True:
    print("=========================")
    print("Escolha a operação:")
    print("=========================")
    print("\n1 - Depósito\n2 - Saque\n3 - Extrato\n4 - Sair")
    opcao = int(input("\nDigite a opção desejada: "))

    if opcao == 1:   # Opção de Depósito
        valor = float(input("\nDigite o valor que quer depositar: "))
        depositar(valor)
    elif opcao == 2: # Opção de Saque
        valor = float(input("Digite o valor que quer sacar: "))
        sacar(valor)
    elif opcao == 3: # Opção de Extrato
        extrato()
    elif opcao == 4: # Opção de Sair do Programa
        break
    else:
        print("\n=========================")
        print("Opção inválida.")
        print("=========================")
```
